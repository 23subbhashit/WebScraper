```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import Select
from tqdm import tqdm
driver = webdriver.Chrome("C:/chromedriver.exe")
driver.maximize_window()
import time
```

```python
driver.get('https://www.jiomart.com/')
```

```python
pincode = 302015 #Enter pincode of your choice    
```

```python
categories = [ "Fruits & Vegetables"
              ,"Dairy & Bakery"
              ,"Staples"
              ,"Snacks & Branded Foods"
              ,"Beverages"
              ,"Personal Care"
              ,"Home Care"
              ,"Baby Care"]
```

```python
driver.find_element_by_name("rel_pincode").send_keys(pincode)
actions = ActionChains(driver)
actions.send_keys(Keys.ENTER).perform()
```

```python
driver.implicitly_wait(10)
```

```python
all_details = []
for category in categories:
    time.sleep(4)
    driver.find_element_by_link_text(category).click()
    driver.execute_script("window.scrollBy(0,500)","")
    
    num = driver.find_element_by_xpath('//*[@id="total_count"]').text
    num = int(num)
    pages = int((num/20)+1)
    print("Number of pages to scrape",pages)
    
    
    
    
    if category == "Fruits & Vegetables":
        category = "fruits-vegetables"
    elif category == "Dairy & Bakery":
        category = "dairy-bakery"
    elif category == "Snacks & Branded Foods":
        category = "snacks-branded-foods"
    elif category == "Personal Care":
        category = "personal-care"
    elif category == "Home Care":
        category = "home-care"
    elif category == "Baby Care":
        category = "baby-care"
    
    
    
    
    #all_details = []
    for page in range(1,pages+1):
        print("Scarping page ",page)
        driver.get('https://www.jiomart.com/category/groceries/{}/page/{}'.format(category.lower(),page))
        e = driver.find_elements_by_class_name("category_name")
        links = []
        for i in e:
            links.append(i.get_property("href"))
        for link in links:
            driver.get(link)
            try:

                r = driver.find_element_by_xpath('//*[@id="center_col"]/div[1]/h1').text
                if category == "fruits-vegetables":
                    p = driver.find_element_by_xpath('//*[@id="price_section"]/div/span[1]/span').text
                else:
                    p = driver.find_element_by_xpath('//*[@id="price_section"]/div/span[2]/span').text
                print(r)
                print(p)
                dicto = {"Name":r,
                        "Price":p}
                all_details.append(dicto)

            except Exception as e:
                print(e)
    driver.get('https://www.jiomart.com/')
```
